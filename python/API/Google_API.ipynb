# Google Places API


# Find by coordinates

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

target_coordinates = "43.6187102, -116.2146068"
target_search = "Chinese"
target_radius = 8000
target_type = "restaurant"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

response = requests.get(base_url, params=params)
places_data = response.json()
print(places_data["results"][0]["name"])
print(places_data["results"][0]["vicinity"])
# China Grand Buffet
# 10498 W Fairview Ave, Boise
pprint(places_data)
```

### Example
- Find the nearest dentist to your house.
- Use Google Maps to find your latitude and Google Places to find
- the dentist. You may also need the rankby property.
- Use geocode to find lat, lng

```python
# Google geocode to find lat, lng
my_address = "151 Sip Ave"

params = {
    "address": my_address,
    "key": gkey
}

geocode_base_url = "https://maps.googleapis.com/maps/api/geocode/json"
my_geo = requests.get(geocode_base_url, params).json()
lat = my_geo["results"][0]["geometry"]["location"]["lat"]
lng = my_geo["results"][0]["geometry"]["location"]["lng"]


# Use lat, lng to use places API to find nearest dentist
target_search = "dentist"

params = {
    "location": f"{lat},{lng}",
    "types": target_search,
    "rankby": "distance",
    "key": gkey
}

# place nearbysearch url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

response = requests.get(base_url, params)
my_dentist = response.json()
print(my_dentist["results"][0]["name"])
print(my_dentist["results"][0]["vicinity"])
```

### Example
- Google place textsearch
- https://maps.googleapis.com/maps/api/place/textsearch/json
- Hint: Read about "Text Search Results"
-(https://developers.google.com/places/web-service/search#TextSearchRequests)
- Find the names and addresses of the top five restaurants in your home city.

```python
my_phrase = "best restaurant in Washington, DC"
target_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"

params = {
    "query": my_phrase,
    "key": gkey
}

response = requests.get(target_url, params)

happy_places = response.json()

counter = 0

for place in happy_places["results"]:
    print(place["name"])
    print(place["formatted_address"])
    counter += 1
    if counter == 5:
        break
```
